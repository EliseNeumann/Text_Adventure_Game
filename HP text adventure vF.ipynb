```python
##########
#
#            DOCTSTRING
#
#    Welcome to the revisited adventure of Harry Potter on his journey to the
#    Philosopher‚Äôs stone!
#
#    This text-based adventure game is inspired by Harry Potter and the
#    Philosopher‚Äôs stone by J.K. Rowling. The goal is to successfully make it
#    through the following obstacles in order to win the game:
#        1. Fluffy‚Äôs room  --  inputting the correct missing lyrics;
#        2. The Chess room  --  determining necessary moves to achieve mate
#           in two;
#        3. The Troll room  --  game of luck.
#
#    Requirements:
#    Images are embedded in the code so entire game file must be downloaded
#    and kept together for the game to run smoothly
#
#    Assumptions:
#        -  General knowledge of American music
#        -  Basic understanding of chess. Though some help is available.
#    While lack of the above does not prevent a player from completing the
#    game, it may put the play time above the 3-minute average.
#
#    Bugs:
#    No bugs have been identified
#
##########


#            ##########################################
#            Importing necessary packages for the game
#            ##########################################
    
import random
from sys import exit
from random import randint
import time
from IPython.display import Image, display



#            #####################
#            Introducing the game
#            #####################

# Titles
print (f"""{"‚ö°Ô∏è„Éª"*5}  HARRRY POTTER AND THE PHILOSOPHER'S STONE  {"„Éª‚ö°Ô∏è"*5}\n\n""")   
print (f"""{"‚òÖ "*10} ‚ú® ‚ú® ‚ú®  Through the trapdoor  ‚ú® ‚ú® ‚ú® {" ‚òÖ"*10}\n\n""")
time.sleep (0.2)     # time.sleep() used occasionally to ensure text appears before prompt

# Asking the player for their name
harry = input (prompt = """Hello and welcome to Hogwarts! What is your name?\n\t""")
harry = harry.capitalize ()

# Narrating introduction
print (f"""
Well {harry}, we are counting on you to make it to the Philosopher's
stone before Voldemort. Good luck getting through all the obstacles you will
encounter on your way!""")

# Moving into the 1st room
input (prompt = "< Press Enter to begin the game > \n")



#            #######################
#            Entering Fluffy's room
#            #######################

# Narrating the start of the room and providing instructions
def fluffy_room () :
    
    print (f"""

{"‚òÖ "*12} ‚ú® ‚ú® ‚ú®  Fluffy's room  ‚ú® ‚ú® ‚ú® {" ‚òÖ"*12}\n""")
    
    print (f"""
You come face to face with a monstrous three-headed dog! The only way to get
past: singing each head to sleep!

Enter only the missing word and note that mispronouncing (aka. mispelling)
equates to a wrong answer, and will only agravate Fluffy!""")
    time.sleep (0.2)
    
    # Creating song 'database'

    queen = """
    üé§ üé∂ Bohemian Rhapsody by Queen
    
    Mama, just killed a man,
    Put a gun against his head,
    Pulled my ____, now he's dead.
    Mama, life had just begun,
    But now I've gone and thrown it all away."""

    whitney = """
    üé§ üé∂ I Wanna Dance With Somebody by Whitney Houston

    Oh! I wanna dance with somebody
    I wanna feel the heat with somebody
    Yeah! I wanna dance with somebody
    With somebody who ____ me"""

    neil = """
    üé§ üé∂ Sweet Caroline by Neil Diamond

    Sweet Caroline
    Good times never seemed so ____
    I've been inclined
    To believe they never would
    Oh, no, no"""
    
    britney = """
    üé§ üé∂ Oops!... I did it again by Britney Spears

    Oops!... I did it again
    I played with your heart, got lost in the ____
    Oh, baby, baby
    Oops!... You think I'm in love
    That I'm sent from above
    I'm not that innocent"""
    
    backstreet = """
    üé§ üé∂ I want it that way by Backstreet Boys

    Tell me why
    Ain't nothin' but a heartache
    Tell me why
    Ain't nothin' but a ____
    Tell me why
    I never wanna hear you say
    I want it that way"""
    
    abba = """
    üé§ üé∂ Dancing Queen by ABBA

    You can dance, you can jive
    Having the time of your life
    Ooh, see that ____, watch that scene
    Digging the dancing queen"""

    beatles = """
    üé§ üé∂ All you need is love by The Beatles

    All you need is love
    All you need is love
    All you need is love, love
    ____ is all you need"""
    
    scott = """
    üé§ üé∂ San Francisco by Scott McKenzie

    If you're going to San Francisco
    Be sure to wear some ____ in your hair
    If you're going to San Francisco
    You're gonna meet some gentle people there"""
    
    jovi = """
    üé§ üé∂ Livin‚Äô on a prayer by Bon Jovi

    Whoa, we're half-way there
    Whoa, livin' on a prayer
    Take my ____, we'll make it. I swear
    Whoa, livin' on a prayer"""

    songs = [queen, whitney, neil, britney, backstreet, abba, beatles, scott, jovi]
    answers = ["trigger", "loves", "good", "game", "mistake", "girl", "love", "flowers", "hand"]

    # Determining the songs player will see
    song_1 = random.choice (songs)
    
    song_2 = random.choice (songs)
    while song_2 == song_1 :
        song_2 = random.choice (songs)
    
    song_3 = random.choice (songs)
    while song_3 == song_1 or song_3 == song_2 :
        song_3 = random.choice (songs)

    
    ## Presenting the first song
    print (f"""Take a deep breath {harry} and start singing!""")
    answer_1 = input (prompt = f"""\t\t {song_1} \n\t""")
    
    # Checking validity of answer
    index_1 = songs.index (song_1)
    
    if answer_1.lower() == answers [index_1] :
        print (f"""
Good job {harry}! Next song:""")
        
    else :
        print (f"""
Fluffy growls at you. You should have sang "{answers [index_1]}". Let's move on to the next one.""")

        
    ## Presenting the second song
    answer_2 = input (prompt = f"""\t\t {song_2} \n\t""")
    
    # Checking validity of answer
    index_2 = songs.index (song_2)
    if answer_2.lower() == answers [index_2] :
        print ("""
That was it! Just one more to go: """)
    else :
        print (f"""
Uh oh. You haven't fooled Fluffy who clearly knew the right answer was
"{answers [index_2]}". Just one more to go: """)

        
    ## Presenting the third song
    answer_3 = input (prompt = f"""\t\t {song_3} \n\t""")
    
    # Checking validity of answer
    index_3 = songs.index (song_3)
    if answer_3.lower() == answers [index_3] :
        print (f"""
You successfully put the third head to sleep!""")
    else :
        print (f"""
Fluffy's third head continues to look at you, unimpressed by your singing.
Maybe if you'd said "{answers [index_3]}" instead of "{answer_3}", things would
be different...""")
        

    ## Calculating and presenting final score
    score = (answer_1 == answers [index_1]) + (answer_2 == answers [index_2]) + (answer_3 == answers [index_3])
    
    # If scored sufficiently high to move on
    if score == 3 :
        input (prompt = """
You got all three heads to sleep! Press Enter to sneak through the trap door
and into the next room. \n""")
        chess_room ()
        
    elif score == 2 :
        input (prompt = """
You got two of the three heads to sleep. That is enough for you to sneak
through the trap door and into the next room by pressing Enter! \n""")
        chess_room ()
    
    # End game if score was too low
    else :
        print (f"""
Fluffy is still very much awake! You cannot get through the trap door so 
unfortunetely your attempt to get to the stone before Voldemort ends here.""")
        fail ()
    
    

#            ########################
#            Entering the Chess room
#            ########################

# Narrating the start of the room and providing instructions
def chess_room ():
    
    print (f"""

{"‚òÖ "*12} ‚ú® ‚ú® ‚ú®  The Chess room  ‚ú® ‚ú® ‚ú® {" ‚òÖ"*12}\n""")
    
    print (f"""
You're standing on the edge of a huge chessboard. On the other side of the
room is another door.
In order to get through, you will need to figure out how to get mate in 2.
This means that two moves are needed for the white team to put the King in
checkmate. 
Because time is short, you cannot afford to make a wrong move!

Best of luck!""")
    
    ready = input (prompt = """
Press Enter to start playing chess, or type "help" to get a brief reminder on
the basic rules of chess \n""")
        
    # Presenting basic chess rules if needed
    if "help" in ready.lower () or "y" in ready.lower ():   
        display(Image(filename = 'chessrules.png'))     # Display rules as a picture       
        input (prompt = """
Hope you're feeling more prepared now! Press Enter to get started \n""")
            
    # Loading pictures in as variables for the game
    PICTURE11 = 'chess11.png'
    PICTURE12 = 'chess12.png'
    PICTURE1S = 'chess1s.png'
    PICTURE21 = 'chess21.png'
    PICTURE22 = 'chess22.png'
    PICTURE2S = 'chess2s.png'
    PICTURE31 = 'chess31.png'
    PICTURE32 = 'chess32.png'
    PICTURE3S = 'chess3s.png'
    
    # Defining the 3 possible games as lists and dictionnaries for later use
        # Below list is constructed as follows : [Picture, possible moves, correct answer]
    games = [[
        [PICTURE11, "D2 to G5, D2 to C3 or D2 to H6", "h6"],
        [PICTURE12, "F6 to E7, F6 to F7 or F6 to G7", "f7"],
    ],
    [
        [PICTURE21, "D1 to G1, D1 to F1 or D1 to D3", "g1"],
        [PICTURE22, "C3 to E4, C3 to E2 or C3 to B5", "e2"],
    ],
    [
        [PICTURE31, "F1 to F4, F1 to D1 or F1 to F8", "f8"],
        [PICTURE32, "H6 to G6, H6 to H8 or H6 to F8", "g6"],
    ]]

    # Defining list of winning images
    win_pics = [PICTURE1S, PICTURE2S, PICTURE3S]

    # Determining the game for this round
    index = random.randint (0,2)
    game = games [index]

    # Executing the 1st round of play
    for picture, move, solution in game :

    # Displaying the chess board for move 1
        print ("""
Here is what the chess board looks like now:""")
        display(Image(filename = picture))
        time.sleep (0.2)

        # Letting player decide what move to make
        print ("""
Where do you need the yellow piece to go to ensure mate in 2?""")
        move = input (prompt = f"""{move} \n""")

        # Checking to see if the player made the right choice
        if solution in move.lower() :
            print (f"""
That was the right move {harry}!""")

        else :
            print (f"""
You were supposed to move to {solution.upper()}. Unfortunately this wrong move
has caused a delay in your journey to the stone and Voldemort has beat you to
it! \n""")
            fail ()
            
    # Printing winning statement
    display(Image(filename = win_pics[index]))
    time.sleep (0.2)
    input (prompt = """
You managed to get checkmate! Press Enter to go through the door and off to
the next room!\n""")
    troll_room()


    
#            ########################    
#            Entering the Troll room
#            ########################

# Narrating the start of the room and providing instructions
def troll_room () :
    
    print (f"""

{"‚òÖ "*12} ‚ú® ‚ú® ‚ú®  The Troll room  ‚ú® ‚ú® ‚ú® {" ‚òÖ"*12}\n""")

   
    print (f"""
A disgusting smell fills your nostrils as your eyes land on the biggest troll
you've ever seen. This is the last obstacle to beating Voldemort to the
Philosopher's stone!

There are two ways for you to attack the Troll:
  -  Stupefy to stun your opponent   : equates to rolling four 12 sided dice
  -  Confringo to blast your opponent: equates to rolling three 18 sided dice

I hope you survive!""")

    # Assigning values to the health vaiables
    harry_hp = 40
    troll_hp = 120

    
    ## While statement so game loops so long as the players still live
    while harry_hp > 0 and troll_hp > 0 :
        print (f"""
        Health points:
        {harry}\t --\t{harry_hp} ‚ô•Ô∏è
        Troll{" " * (len(harry) - 5)}\t --\t{troll_hp} ‚ô•Ô∏è""")   # Formatting with len for better visuals
        time.sleep (0.2)
        
        # Defining weapons (variables) available
        stupefy = {"name" : "Stupefy" , "hit" : (randint (1,12)) + (randint (1,12)) + (randint (1,12)) + (randint (1,12))}
        confringo = {"name" : "Confringo" , "hit" :(randint (1,18)) + (randint (1,18)) + (randint (1,18))}
        troll_club = {"name" : "club" , "hit" :(randint (1,9)) + (randint (1,9))}
        troll_fists = {"name" : "fists" , "hit" :randint (1,20)}
        troll_weapons = (troll_club, troll_fists)
        
        # Selecting the weapons for each player - a random value and a user input
        weapon_troll = random.choice (troll_weapons)
        weapon_hp = input (prompt = """
What spell will you use this round, Stupefy or Confringo? \t""")

        
        ## Adding trick question when player's life gets low
        if harry_hp < 12 :
            weapon_hp = input (prompt = """
Just before you utter the spell, you realize how weak you've gotten from all
the hits you took... In your dazed state of mind, you remember that the Avada
Kadavra curse could probably finish off the troll... But it is one of the
unforgivable curses... What to do?! Rethink your spell choice for this round
and make a decision:
Avada Kadavra, Stupefy, or Confringo? \t""")
            
            # End game if player decided to use Avada Kadavra
            if "ava" in weapon_hp.lower () or "1" in weapon_hp.lower ():
                harry_hp = harry_hp - 12
                print (f"""
You should have known better than to do that, {harry}. Using the 
unforgivable curses means a long sentence in the wizard prison: Azkaban.
Dementors immediately appear around you to cart you off. \n""")
                fail ()
            
            else :
                
                # Determining damage done by each player with stupefy spell
                if "stu" in weapon_hp.lower () or "2" in weapon_hp.lower () :
                    troll_hp -= stupefy ["hit"]
                    harry_hp -= weapon_troll ["hit"]
            
                    # Narrating what just happened
                    print (f"""
        Damage done by each player:
        {harry} causes {stupefy ["hit"]} damage to the Troll with Stupefy
        The Troll causes {weapon_troll ["hit"]} damage to {harry} with his {weapon_troll["name"]}""")
        
                # Determining damage done by each player with confringo spell
                elif "con" in weapon_hp.lower () or "3" in weapon_hp.lower () :
                    troll_hp -= confringo ["hit"]
                    harry_hp -= weapon_troll ["hit"]

                    # Narrating what just happened
                    print (f"""
        Damage done by each player:
        {harry} causes {confringo ["hit"]} damage to the Troll with Confringo
        The Troll causes {weapon_troll ["hit"]} damage to {harry} with his {weapon_troll["name"]}""")

                # Else as catch-all for issues
                else :
                    print ("""
I'm sorry but if you don't properly articulate when saying spells, you might
as well not say anything at all. Let's go through it again.\n""")
        
        # Determining damage done by each player 
        else :
            
             # Determining damage done by each player with stupefy spell
            if "stu" in weapon_hp.lower () or "1" in weapon_hp.lower ():
                troll_hp -= stupefy ["hit"]
                harry_hp -= weapon_troll ["hit"]

            # Narrating what just happened
                print (f"""
        Damage done by each player:
        {harry} causes {stupefy ["hit"]} damage to the Troll with Stupefy
        The Troll causes {weapon_troll ["hit"]} damage to {harry} with his {weapon_troll["name"]}""")

            # Determining damage done by each player with confringo spell
            elif "con" in weapon_hp.lower () or "2" in weapon_hp.lower () :
                troll_hp -= confringo ["hit"]
                harry_hp -= weapon_troll ["hit"]

                # Narrating what just happened
                print (f"""
        Damage done by each player:
        {harry} causes {confringo ["hit"]} damage to the Troll with Confringo
        The Troll causes {weapon_troll ["hit"]} damage to {harry} with his {weapon_troll["name"]}""")

            # Else as catch-all for issues
            else :
                print ("""
I'm sorry but if you don't properly articulate when saying spells, you might
as well not say anything at all. Let's go through it again.\n""")

    # If both die
    if harry_hp <= 0 and troll_hp <= 0 :
        print ("""
Both you and the troll dealt killing blows. That is a win for no one...""")
        fail ()
        
    # If Harry dies
    elif harry_hp <= 0 :
        print ("""
Unfortunately the troll got to you before you could get to it!""")
        fail ()

    # If troll dies
    elif troll_hp <= 0 :
        print (f"""
You killed the troll! Great job {harry}! \n""")
        
        # Offering the player and oportunity to play again
        print (f"""

{" " * 20}üçæ ‚ú® üçæ ‚ú® üçæ ‚ú®  WIN  ‚ú® üçæ ‚ú® üçæ ‚ú® üçæ{" " * 20}""")
        print ("""
You successfully prevented Voldemort from getting the Philosopher's stone,
and have, at least for now, delayed his return... \n\n""")
        input (prompt = "Press Enter to play again \n")
        fluffy_room ()
        
    # Catch-all else statement
    else :
        print ("""
Something went wrong... it must be your faut. I'm feeling generous though so
let's give you one more shot at this fight.""")
        troll_room ()

#            ######################## 
#            Defining the lose option
#            ######################## 

def fail () :
    print (f"""\n\n{" " * 18}üö´ ‚ò†Ô∏è üö´ ‚ò†Ô∏è üö´ ‚ò†Ô∏è  GAME OVER  ‚ò†Ô∏è üö´ ‚ò†Ô∏è üö´ ‚ò†Ô∏è üö´{" "*18}\n""")
    time.sleep (0.2)
    input (prompt = """Press Enter to try again\n""")
    print (f"""
All right {harry}, hang onto your broomstick cause here we go again! \n\n""")
    fluffy_room ()
    
    
    
fluffy_room()

```
